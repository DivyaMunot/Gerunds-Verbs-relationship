```python
# import the required libraries

import pandas as pd
import nltk
from nltk.corpus import verbnet
import spacy
from spacy import displacy
from nltk.stem import PorterStemmer

nltk.download("punkt")
nlp = spacy.load("en_core_web_sm")
nltk.download('verbnet')
ps = PorterStemmer()
```

```python
# reading the dataframe 
df=pd.read_excel('/content/dataframe.xlsx')
```

```python
# Finding the verb class
df["depending_on_verb"] = df["depending_on_verb"].str.strip("[]").str.split(", ")
l=[]
main_list_verb = []
h=df["depending_on_verb"]
s=df["sentence"]
j=-1
for i,row in df.iterrows():
  sentence = row['sentence']
  list_verb = []
  sentence=str(sentence)
  doc = nlp(sentence)
  list_ = []
  counter=1
  for token in doc:
      if token.pos_ == "VERB" and token.text in h[i]:
        vnclass = verbnet.classids(lemma = token.lemma_)
        counter=0
        if vnclass:
          l.append(vnclass)
        else:
          l.append([""])
```

```python
# saving the verb_type
df['verb_class'] = l
df.to_csv("dataframe_verb_class.csv")
```

```python
# finding the verb class type
g=df1["verb_class"]
class_type=[]
for i in g:
  print(i[0])
  if i[0]!="''":
    num = i[0].split("-")[1]
    num=num.split('.')
    num=num[0]
    if num[-1]=="'":
      num=num[:-1]
    num=int(num)
    print(vnclass2vntype[num])
    class_type.append(vnclass2vntype[num])
  else:
    class_type.append("None")
```
